## Missing Data

In the Towards Data Science article, they list 3 techniques for detecting any missing data frame. Try all three before deciding on a way to solve any missing data issues.

```python
# Plot a heatmap of any missing data
cols = womens_clothing_data.columns[:11]
colours = ["#000099", "#ffff00"]
sns.heatmap(womens_clothing_data[cols].isnull(), cmap = sns.color_palette(colours))
```

```python
# Try out a missing data percentage list! 
for col in womens_clothing_data.columns:
    pct_missing = np.mean(womens_clothing_data[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Plot a missing data histogram
for col in womens_clothing_data.columns:
    missing = womens_clothing_data[col].isnull()
    num_missing = np.sum(missing)
    
    if num_missing >0:
        print('created missing indicator for: {}'.format(col))
        womens_clothing_data['{}_ismissing'.format(col)] = missing
        
ismissing_cols = [col for col in womens_clothing_data.columns if 'ismissing' in col]
womens_clothing_data['num_missing'] = womens_clothing_data[ismissing_cols].sum(axis = 1)

womens_clothing_data['num_missing'].value_counts().reset_index().sort_values(by='index').plot.bar(x='index', y = 'num_missing')
```
